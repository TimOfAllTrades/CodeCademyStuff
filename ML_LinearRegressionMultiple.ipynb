```python
import matplotlib.pyplot as plt
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

streeteasy = pd.read_csv("sampledata/manhattan.csv")

#Import the data csv
df = pd.DataFrame(streeteasy)

#This are the values which we want to use to predict the rent
x = df[['bedrooms', 'bathrooms', 'size_sqft', 'min_to_subway', 'floor', 'building_age_yrs', 'no_fee', 'has_roofdeck', 'has_washer_dryer', 'has_doorman', 'has_elevator', 'has_dishwasher', 'has_patio', 'has_gym']]

#Rent is what we want to predict, so that is the y value
y = df[['rent']]

#Splitting the whole data into a training set and test set
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.8, test_size = 0.2, random_state=6)

#Instantiate the linear regression model
lm = LinearRegression()

#Fit the data using a linear regression and the x and y training data
model = lm.fit(x_train, y_train)

#Predict the labels or results using the x training data
y_predict= lm.predict(x_test)

#Print the score of the training data
print("Train score:")
print(lm.score(x_train, y_train))

#Print the score of the test data
print("Test score:")
print(lm.score(x_test, y_test))

#Plot the predicted prices vs the actual prices.  Should be a diagonal line
plt.scatter(y_test, y_predict, alpha=0.4)
plt.plot(range(20000), range(20000))

plt.xlabel("Prices: $Y_i$")
plt.ylabel("Predicted prices: $\hat{Y}_i$")
plt.title("Actual Rent vs Predicted Rent")

plt.show()
```
